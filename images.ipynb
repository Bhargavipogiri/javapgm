```python
import os

os.makedirs("dogs", exist_ok=True)
```

```python
import requests

dog_image_urls = [
    "https://images.unsplash.com/photo-1558788353-f76d92427f16",
    "https://images.unsplash.com/photo-1517849845537-4d257902454a",
    "https://images.unsplash.com/photo-1507146426996-ef05306b995a",
    "https://images.unsplash.com/photo-1517423440428-a5a00ad493e8",
    "https://images.unsplash.com/photo-1525253086316-d0c936c814f8"
]

for i, url in enumerate(dog_image_urls, start=1):
    response = requests.get(url)
    with open(f"dogs/dog_{i}.jpg", "wb") as f:
        f.write(response.content)

print("Dog images downloaded successfully")
```

```python
import shutil

shutil.make_archive("dogs_images", "zip", "dogs")
print("dogs_images.zip created")
```
